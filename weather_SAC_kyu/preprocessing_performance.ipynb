```python
import pandas as pd
```

# 공연입장객 데이터 로드

```python
performance = pd.read_excel('./data/공연입장객/예술의전당_공연 및 전시 입장객 현황_20251231.xlsx', sheet_name=['2023년(공연)', '2023년(전시)', '2024년(공연)', '2024년(전시)', '2025년(공연)', '2025년(전시)'])
final_performance = []
for i in range(3,6):
    final_performance.append(performance[f'202{i}년(공연)'])
    final_performance.append(performance[f'202{i}년(전시)'])
    
final_performance = pd.concat(final_performance)
final_performance = final_performance[['일자', '합계']]
final_performance = final_performance.rename(columns={'일자': '날짜'})
final_performance
```

```python
# 날짜 형식 맞추기
final_performance['날짜'] = final_performance['날짜'].astype(str).str.split('.').str[0]

final_performance['날짜'] = pd.to_datetime(final_performance['날짜'], format='mixed')

final_performance = final_performance.groupby('날짜')['합계'].sum().reset_index()

final_performance
```
